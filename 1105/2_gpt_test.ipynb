```python
from dotenv import load_dotenv

#.env 파일에서 OPENAI_API_KEY 이름의 키값을 가져옵니다.
# -> OPENAI() 객체 생성할 때 사용합니다.
load_dotenv()
```

```python
from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "system",
            "content": (
                "당신은 mbti가 t인 공감능력 결여자입니다"
                "무슨 말을 하든 현실적인 답변을 주세요"
                "대문자 t인 거에요"
                "상대방 말에 팩폭을 날리세요"
            )
        },
        {
            "role": "user",
            "content": (
                "나 우울해서 빵 샀어..."
                "여자친구가 너무 보고 싶어..."
            )
        }
    ],
    temperature=0.7,  
    # 출력의 창의성과 무작위성을 조절하는 파라미터.(0~1  :사용하는 단어 일반적~창의적)
    max_tokens=300
)
print(response.model_dump_json(indent=2))
# print(response.choices[0].message["content"])
```

```python
print(f"ai: {response.choices[0].message.content}")
```

```python
response = client.responses.create(
  model="gpt-5-nano",
  input="연애하는 법"
)

print(response.model_dump_json(indent=2))
print(f'🤖 AI : {response.output_text}')
```
